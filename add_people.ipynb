```python
def add_people():
    group_name = get_group_name()
    number_of_members = group_size_confirm()
    people = []

    for i in range(number_of_members):
        print(f"\nEntering person {i + 1} of {number_of_members}")
        person = get_person_name()
        people.append(person)

    print("\n--- Group Summary ---")
    print(f"Group Name (Location): {group_name}")
    print("Members:")
    for person in people:
        print(person)

    return group_name, people
```
