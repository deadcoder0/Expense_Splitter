```python
def get_group_name():
    while True:
        group_name = input("Enter your group name (location): ").strip()

        if len(group_name) == 0:
            print("Error! Group name cannot be empty!")
            continue

        if len(group_name) > 20:
            print("Error! Max 20 characters allowed!")
            continue

        # Optional: allow letters + spaces only
        if not all(c.isalpha() or c.isspace() for c in group_name):
            print("Error! Only letters and spaces allowed!")
            continue

        return group_name
```
